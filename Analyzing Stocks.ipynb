#**GATHERING STOCK DATA**

```python
# Importing libraries
"""!pip install yfinance
!pip install praw
!pip install bs4
!pip install nbformat
!pip install --upgrade plotly"""
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
from bs4 import BeautifulSoup
import requests
```

```python
# Getting and storing ticker data  for the Producers

netflix = yf.Ticker("NFLX")
netflix_data = netflix.history(period = "max")
netflix_data.reset_index(inplace = True)

warner_bros = yf.Ticker("WBA")
warner_bros_data = warner_bros.history(period = "max")
warner_bros_data.reset_index(inplace = True)

disney = yf.Ticker("DIS")
disney_data = disney.history(period = "max")
disney_data.reset_index(inplace = True)

kartoon_studios = yf.Ticker("TOON")
kartoon_studios_data = kartoon_studios.history(period = "max")
kartoon_studios_data.reset_index(inplace = True)

Sony = yf.Ticker("SONY")
Sony_data = Sony.history(period = "max")
Sony_data.reset_index(inplace = True)


# Getting and storing ticker data for the Distributors
funko = yf.Ticker("FNKO")
funko_data = funko.history(period = "max")
funko_data.reset_index(inplace = True)

gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period = "max")
gamestop_data.reset_index(inplace = True)


hasbro = yf.Ticker("HAS")
hasbro_data = hasbro.history(period = "max")
hasbro_data.reset_index(inplace = True)


# Filter data for COVID-19 period

start_date = '2020-01-11'
end_date = '2023-05-11'

netflix_filtered = netflix_data[(netflix_data['Date'] >= start_date) & (netflix_data['Date'] <= end_date)]
warner_bros_filtered = warner_bros_data[(warner_bros_data['Date'] >= start_date) & (warner_bros_data['Date'] <= end_date)]
disney_filtered = disney_data[(disney_data['Date'] >= start_date) & (disney_data['Date'] <= end_date)]
funko_filtered = funko_data[(funko_data['Date'] >= start_date) & (funko_data['Date'] <= end_date)]
kartoon_studios_filtered = kartoon_studios_data[(kartoon_studios_data['Date'] >= start_date) & (kartoon_studios_data['Date'] <= end_date)]
gamestop_filtered = gamestop_data[(gamestop_data['Date'] >= start_date) & (gamestop_data['Date'] <= end_date)]
hasbro_filtered = hasbro_data[(hasbro_data['Date'] >= start_date) & (hasbro_data['Date'] <= end_date)]
Sony_filtered = Sony_data[(Sony_data['Date'] >= start_date) & (Sony_data['Date'] <= end_date)]
```

```python
from logging import warn
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

url_list=["https://companiesmarketcap.com/netflix/earnings/", "https://companiesmarketcap.com/warner-bros-discovery/earnings/",
"https://companiesmarketcap.com/walt-disney/earnings/","https://companiesmarketcap.com/funko/earnings/", "https://companiesmarketcap.com/kartoon-studios/earnings/",
"https://companiesmarketcap.com/gamestop/earnings/", "https://companiesmarketcap.com/hasbro/earnings/", "https://companiesmarketcap.com/sony/earnings/"]

#Netflix
netflix_url = url_list[0]

response = requests.get(netflix_url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
table = pd.read_html(soup.find('table').prettify())
netflix_revenue = table[0]

netflix_earnings = pd.DataFrame(netflix_revenue)
netflix_earnings.to_csv("netflix_earnings.csv")

#Warner Bros
warner_bros_url = url_list[1]

response = requests.get(warner_bros_url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
table = pd.read_html(soup.find('table').prettify())
warner_bros_revenue = table[0]

warner_bros_earnings = pd.DataFrame(warner_bros_revenue)
warner_bros_earnings.to_csv("warner_bros_earnings.csv")

#Disney
disney_url = url_list[2]

response = requests.get(disney_url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
table = pd.read_html(soup.find('table').prettify())
disney_revenue = table[0]

disney_earnings = pd.DataFrame(disney_revenue)
disney_earnings.to_csv("disney_earnings.csv")

#Funko
funko_url = url_list[3]

response = requests.get(funko_url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
table = pd.read_html(soup.find('table').prettify())
funko_revenue = table[0]

funko_earnings = pd.DataFrame(funko_revenue)
funko_earnings.to_csv("funko_earnings.csv")


#Kartoon Studios
kartoon_studios_url = url_list[4]

response = requests.get(kartoon_studios_url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
table = pd.read_html(soup.find('table').prettify())
kartoon_studios_revenue = table[0]

kartoon_studios_earnings = pd.DataFrame(kartoon_studios_revenue)
kartoon_studios_earnings.to_csv("kartoon_studios_earnings.csv")


#GameStop
gamestop_url = url_list[5]

response = requests.get(gamestop_url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
table = pd.read_html(soup.find('table').prettify())
gamestop_revenue = table[0]

gamestop_earnings = pd.DataFrame(gamestop_revenue)
gamestop_earnings.to_csv("gamestop_earnings.csv")


#Hasbro
hasbro_url = url_list[6]

response = requests.get(hasbro_url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
table = pd.read_html(soup.find('table').prettify())
hasbro_revenue = table[0]

hasbro_earnings = pd.DataFrame(hasbro_revenue)
hasbro_earnings.to_csv("hasbro_earnings.csv")


#Sony
sony_url = url_list[7]

response = requests.get(sony_url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
table = pd.read_html(soup.find('table').prettify())
Sony_revenue = table[0]

Sony_earnings = pd.DataFrame(Sony_revenue)
Sony_earnings.to_csv("Sony_earnings.csv")


```

Using ARIMA, analyze the COVID-19 opening and closing stock prices for each company and create a new forecast based on the events of a new recession.

##Producers

```python
from prophet import Prophet

# Prepare data for Prophet
df = netflix_filtered[['Date', 'Close']].rename(columns={'Date': 'ds', 'Close': 'y'})

# Remove timezone
df['ds'] = pd.to_datetime(df['ds']).dt.tz_localize(None)

# Fit model
model = Prophet(daily_seasonality=True)
model.fit(df)

# Make future dates
forecast_period = 750
future = model.make_future_dataframe(periods=forecast_period)
future['ds'] = pd.to_datetime(future['ds']).dt.tz_localize(None)

# Forecast
forecast = model.predict(future)

# Plot
plt.figure(figsize=(15,7))
plt.plot(df['ds'], df['y'], label="Netflix Close Price")
plt.plot(forecast['ds'], forecast['yhat'], label="Forecasted Price", linestyle="--", color="red")
plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'],
                 color="pink", alpha=0.3, label="Confidence Interval")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Netflix Stock Price Forecast (Prophet)")
plt.legend()
plt.grid(True)
plt.show()
```
