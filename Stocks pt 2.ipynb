```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Getting and storing ticker data for the lifestyle companies

nike = yf.Ticker("NKE")
nike_data = nike.history(period = "max")
nike_data.reset_index(inplace = True)

lululemon = yf.Ticker("LULU")
lululemon_data = lululemon.history(period = "max")
lululemon_data.reset_index(inplace = True)

peloton = yf.Ticker("PTON")
peloton_data = peloton.history(period = "max")
peloton_data.reset_index(inplace = True)

sprouts = yf.Ticker("SFM")
sprouts_data = sprouts.history(period = "max")
sprouts_data.reset_index(inplace = True)

home_depot = yf.Ticker("HD")
home_depot_data = home_depot.history(period = "max")
home_depot_data.reset_index(inplace = True)

starbucks = yf.Ticker("SBUX")
starbucks_data = starbucks.history(period = "max")
starbucks_data.reset_index(inplace = True)


# Filter data for COVID-19 period

start_date = '2020-01-11'
end_date = '2023-05-11'

nike_filtered = nike_data[(nike_data['Date'] >= start_date) & (nike_data['Date'] <= end_date)]
lululemon_filtered = lululemon_data[(lululemon_data['Date'] >= start_date) & (lululemon_data['Date'] <= end_date)]
peloton_filtered = peloton_data[(peloton_data['Date'] >= start_date) & (peloton_data['Date'] <= end_date)]
sprouts_filtered = sprouts_data[(sprouts_data['Date'] >= start_date) & (sprouts_data['Date'] <= end_date)]
home_depot_filtered = home_depot_data[(home_depot_data['Date'] >= start_date) & (home_depot_data['Date'] <= end_date)]
starbucks_filtered = starbucks_data[(starbucks_data['Date'] >= start_date) & (starbucks_data['Date'] <= end_date)]
```

```python
# Lifestyle Companies

# Nike
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(nike_data['Date'], nike_data['Close'])
axes[0].set_title("Nike Stock Price (Max Period)")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Close Price")

axes[1].plot(nike_filtered['Date'], nike_filtered['Close'])
axes[1].set_title("Nike Stock Price During COVID")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Close Price")
plt.tight_layout()
plt.show()

# Lululemon
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(lululemon_data['Date'], lululemon_data['Close'])
axes[0].set_title("Lululemon Stock Price (Max Period)")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Close Price")

axes[1].plot(lululemon_filtered['Date'], lululemon_filtered['Close'])
axes[1].set_title("Lululemon Stock Price During COVID")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Close Price")
plt.tight_layout()
plt.show()

# Peloton
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(peloton_data['Date'], peloton_data['Close'])
axes[0].set_title("Peloton Stock Price (Max Period)")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Close Price")

axes[1].plot(peloton_filtered['Date'], peloton_filtered['Close'])
axes[1].set_title("Peloton Stock Price During COVID")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Close Price")
plt.tight_layout()
plt.show()

# Sprouts Farmers Market
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(sprouts_data['Date'], sprouts_data['Close'])
axes[0].set_title("Sprouts Farmers Market Stock Price (Max Period)")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Close Price")

axes[1].plot(sprouts_filtered['Date'], sprouts_filtered['Close'])
axes[1].set_title("Sprouts Farmers Market Stock Price During COVID")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Close Price")
plt.tight_layout()
plt.show()

# Home Depot
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(home_depot_data['Date'], home_depot_data['Close'])
axes[0].set_title("Home Depot Stock Price (Max Period)")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Close Price")

axes[1].plot(home_depot_filtered['Date'], home_depot_filtered['Close'])
axes[1].set_title("Home Depot Stock Price During COVID")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Close Price")
plt.tight_layout()
plt.show()

# Starbucks
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(starbucks_data['Date'], starbucks_data['Close'])
axes[0].set_title("Starbucks Stock Price (Max Period)")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Close Price")

axes[1].plot(starbucks_filtered['Date'], starbucks_filtered['Close'])
axes[1].set_title("Starbucks Stock Price During COVID")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Close Price")
plt.tight_layout()
plt.show()
```
